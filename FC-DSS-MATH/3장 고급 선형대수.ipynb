# 3.5 PCA

```python
from sklearn.datasets import load_iris

iris = load_iris()
N = 10
X = iris.data[:N, :2]
```

```python
type(X)
```

```python
plt.plot(X.T, 'o:') # plot은 ndarray의 열을 하나의 데이터 모임으로 본다
plt.xticks(range(2), ['꽃받침 길이', '꽃받침 폭'])
plt.xlim(-0.5, 2)
plt.ylim(2.5, 6)
plt.title('붓꽃 크기 특성')
plt.legend([f'표본 {i+1}' for i in range(N)])
plt.show()
```
